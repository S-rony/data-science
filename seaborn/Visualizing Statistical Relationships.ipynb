```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='total_bill', y = 'tip', data=df, kind = 'scatter')
plt.show()
```

```python
sns.relplot(x='total_bill', y = 'tip', data=df, hue = 'time')
plt.show()
```

```python
sns.relplot(x='total_bill', y = 'tip', data=df, style = 'time', hue = 'smoker')
plt.show()
```

```python
sns.relplot(x='total_bill', y = 'tip', data=df, hue = 'time', col = 'smoker')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y = 'tip', data=df, hue = 'time')
plt.show()
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.relplot(x='mpg', y = 'horsepower', data = df, kind = 'scatter', col='origin')

plt.show()
print("USA, has the majority data point")
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
sns.lineplot(x= 'year', y = 'passengers', data = df,hue = 'month')
plt.show()
```

```python
sns.relplot(x= 'year', y = 'passengers',kind = 'line', data = df, col = 'month', hue= 'month')
plt.show()
```
