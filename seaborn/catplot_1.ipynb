```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x='sex', y = 'total_bill', data= df )
plt.show()
```

```python
sns.catplot(x='sex', y = 'total_bill', data= df, jitter=1 )
plt.show()
```

```python
sns.catplot(x='sex', y = 'total_bill', data= df, jitter=0.75 )
plt.show()
```

```python
sns.catplot(x='sex', y = 'total_bill', data= df, jitter=0.25 )
plt.show()
```

```python
sns.catplot(x='sex', y = 'total_bill', data= df, jitter=0.005, hue = 'smoker' )
plt.show()
```

```python
 sns.catplot(x='sex', y = 'total_bill', data= df, kind= 'swarm' )
plt.show()
```

```python
 sns.catplot(x='sex', y = 'total_bill', data= df, hue='smoker',kind= 'swarm' )
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data= df, kind = 'box')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data= df, kind = 'box',hue= 'smoker')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data= df, kind = 'boxen',hue= 'smoker')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data= df, kind = 'violin',hue= 'smoker')
plt.show()
```
